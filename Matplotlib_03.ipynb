```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Dept': [
        'HR', 'IT', 'Sales', 'IT', 'Sales', 'HR', 'IT', 'Marketing', 'Sales', 'HR',
        'Marketing', 'IT', 'HR', 'Sales', 'Marketing', 'IT', 'HR', 'Sales', 'Marketing', 'IT'
    ],
    'Salary': [
        50000, 62000, 45000, 70000, 48000, 52000, 65000, 55000, 47000, 51000,
        58000, 66000, 53000, 49000, 56000, 68000, 54000, 46000, 57000, 69000
    ],
    'Age': [
        25, 30, 28, 35, 24, 29, 32, 27, 26, 31,
        34, 33, 28, 25, 30, 36, 29, 24, 33, 35
    ],
    'Experience': [
        2, 5, 4, 10, 1, 4, 7, 3, 2, 6,
        9, 8, 4, 2, 5, 11, 5, 1, 8, 10
    ]
}

df = pd.DataFrame(data)
df.head()
```

# Multivariate analysis

```python
sorted_df = df[['Salary', 'Experience']].sort_values(by='Salary')
```

```python
# Bubble plot using Age (x), Salary (y), and Experience (size)
plt.figure(figsize=(7, 5))
plt.scatter(df['Age'], df['Salary'], s=df['Experience']*30, color='red')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary vs Age')
plt.grid(True)
plt.show()
```

# 2 numerical and 1 categorical

```python
plt.figure(figsize=(9, 6))

for dept in df['Dept'].unique():
    subset = df[df['Dept'] == dept]
    plt.scatter(subset['Age'], subset['Salary'], label=dept, s=100, alpha=0.7)

plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary vs Age by Department (Matplotlib Only)')
plt.legend(title='Department')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

#  Object oriented programming

```python
fig,axs = plt.subplots(1,3 , figsize=(15,5))

#  Line plot:

axs[0].plot(df["Age"],df["Salary"],color="red",marker="*",markersize=2,linewidth="2")
axs[0].grid()
axs[0].set_title("Line plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")

# Histogram:

axs[1].hist(df["Salary"],bins=7,color="skyblue")
axs[1].grid()
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")

# Boxplot:

axs[2].boxplot(df["Salary"])
axs[2].grid()
axs[2].set_title("Boxplot")
axs[2].set_xlabel("Salary")


plt.show()
```

```python
df_sorted_age = df.sort_values('Age')

fig, axs = plt.subplots(1, 3, figsize=(18, 5))
#  Line Plot:
axs[0].plot(df_sorted_age['Age'], df_sorted_age['Salary'], color='blue', marker='o')
axs[0].set_title('Salary Trend by Age (Sorted)')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Salary')
axs[0].grid()

# Histogram:
axs[1].hist(df['Salary'], bins=8, color='lightgreen', edgecolor='black')
axs[1].set_title('Salary Distribution')
axs[1].set_xlabel('Salary')

# Boxplot:
axs[2].boxplot(df['Salary'], vert=False, patch_artist=True)
axs[2].set_title('Salary Range & Median')
axs[2].set_xlabel('Salary')

# TO save image

# plt.savefig('salary_analysis.png', dpi=300, bbox_inches='tight')

plt.tight_layout()
plt.show()
```

```python
data= {
    'Year':[2010,2011,2012,2013],
    'sales':[100,110,100,130],
    'profit':[10,20,30,40],
    'expenses':[40,20,60,90]
}
df = pd.DataFrame(data)
df.head()
```

```python
plt.plot(df["Year"],df["sales"],label="sales")
plt.plot(df["Year"],df["profit"],label="profit")
plt.plot(df["Year"],df["expenses"],label="expenses")
plt.legend()
plt.show()
```

# 3-D plot

```python
ax = plt.axes(projection='3d')
# Using df_sorted_age which contains the employee data
ax.scatter(df_sorted_age['Age'], df_sorted_age['Salary'], df_sorted_age['Experience'])
ax.set_xlabel('Age')
ax.set_ylabel('Salary')
ax.set_zlabel('Experience')
plt.title(' Age, Salary, and Experience')
plt.show()
```
