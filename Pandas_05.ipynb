```python
import pandas as pd
```

```python
data={
    'Name':['Alice','Bob','Claire','syam','karthik','siddhu','mahesh','fardeen','teja','saaqib'],
    'Gender':['Female','Male','Female','Male','Male','Female','Male','Male','Male','Male'],
    'grade':['A','B','A','A','c','B','A','B','A','B'],
    'marks':[25,30,27,22,65,47,88,79,66,92]
}
df=pd.DataFrame(data)
df
```

```python
df.groupby('grade').count()
```

```python
df.groupby('grade')['marks'].mean()
```

```python
df.groupby('grade')['marks'].max()
```

```python

df.groupby('Gender')['marks'].agg(['mean','median','min','max','count','sum'])
```

```python
df.groupby('Gender')['marks'].agg(Average='mean',median='median',lowest='min',highest='max')
```

```python
reg1=pd.DataFrame({'id':[1,2,3],'name':['Manoj','gsfejh','Arham']})
reg2=pd.DataFrame({'id':[1,2,3],'marks':[96,44,33]})

pd.merge(reg1,reg2,on='id')
```

```python
reg1=pd.DataFrame({'id':[1,2,3],'name':['Manoj','gsfejh','Arham']})
reg2=pd.DataFrame({'id':[1,2,3],'marks':[96,44,39]})

pd.merge(reg1,reg2,on='id',how='inner')
```

```python
reg1=pd.DataFrame({'id':[1,2,3],'name':['Manoj','gsfejh','Arham']})
reg2=pd.DataFrame({'id':[1,2,3],'marks':[96,44,39]})

pd.merge(reg1,reg2,on='id',how='left')
```

```python
reg1=pd.DataFrame({'id':[1,2,3],'name':['Manoj','gsfejh','Arham']})
reg2=pd.DataFrame({'id':[1,2,3],'marks':[96,44,39]})

pd.concat([reg1,reg2])
```

```python
reg1=pd.DataFrame({'id':[1,2,3],'name':['Manoj','gsfejh','Arham']})
reg2=pd.DataFrame({'c':[81,24,53],'marks':[96,44,39]})

pd.concat([reg1,reg2],axis=1)
```
