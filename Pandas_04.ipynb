### Find columns that contain letters
To find columns that contain letters, we can first filter for `object` type columns (which typically hold strings) and then check each value in those columns for the presence of alphabetic characters using regular expressions.

```python
columns_with_letters = []
for col in df.select_dtypes(include='object').columns:
    # Check if any value in the column contains at least one letter
    if df[col].astype(str).str.contains('[a-zA-Z]').any():
        columns_with_letters.append(col)

print("Columns containing letters:", columns_with_letters)
```
