### Bivariate Analysis: Visualizing Two Columns
Below are examples of how to use Histograms and Boxplots to compare two different data series.

```python
import numpy as np

# Create a second column for comparison (e.g., 'y' values)
df['y'] = df['x'] * 0.8 + np.random.normal(0, 1000, len(df))

# Example 1: Overlaid Histograms
plt.figure(figsize=(8, 5))
plt.hist(df['x'], bins=10, alpha=0.5, label='Series X', color='blue')
plt.hist(df['y'], bins=10, alpha=0.5, label='Series Y', color='orange')
plt.title('Overlaid Histogram: Comparing Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Example 2: Side-by-Side Boxplots
plt.figure(figsize=(8, 5))
# Updated 'labels' to 'tick_labels' to address the deprecation warning
plt.boxplot([df['x'], df['y']], tick_labels=['Series X', 'Series Y'])
plt.title('Side-by-Side Boxplot: Comparing Summaries')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np

# Create some dummy NumPy data
features = np.linspace(0, 10,100)
predictions = features ** 2

# Create a Figure and one Axes object
fig, ax = plt.subplots()

ax.plot(features, predictions, color='red', label='Model Prediction')
ax.set_title("Feature vs Prediction")
ax.set_xlabel("Input Feature")
ax.set_ylabel("Target Variable")
ax.legend() # Crucial for ML to show what lines mean what!

plt.show()
```

```python
# Creates a 1x2 grid of plots (1 row, 2 columns)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plot on the left graph
ax1.scatter(x, y, color='blue')
ax1.set_title("Raw Data Scatter")

# Plot on the right graph
ax2.plot(x, y, color='green', linestyle='--')
ax2.set_title("Trend Line")

plt.tight_layout() # Prevents overlapping labels
plt.show()
```
